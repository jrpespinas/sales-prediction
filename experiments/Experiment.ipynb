### 1. Handling Missing Values

There are three reasons why a dataset consists of missing values:
1. Missing not at random (MCAR): the value itself may be the reason why it is missing
2. Missing at random (MAR): the value depends on another value--one feature influences another feature
3. Missing completely at random (MCAR): there is no particular reason why it is missing, however it's a rare case

There are two ways to deal with missing values:
1. Imputation: fill missing values with default, risky for differentiating MAR/MCAR or 0 value.
2. Deletion
    1. Row deletion: only good for MCAR and if the number of MCAR cases are less than 0.1%, risky for MAR cases, introduces bias
    2. Column deletion: risky for MAR features

So let us observe for **MCAR** and **MAR** values.

---

```python
# identify columns with NaN values
df.isnull().sum()
```

#### Promo2 data

Since **Promo2** dictates whether the store participates in a continuous promotion run, **Promo2SinceWeek**, **Promo2SinceYear**, and **PromoInterval** should be imputed instead of deleting as this a case of MAR.

We will use 0 as the default value for NaN of the said columns

```python
cols = ['Promo2', 'Promo2SinceWeek','Promo2SinceYear', 'PromoInterval']
promos = df[cols]
promos
```

```python
# define the values to replace NaN per column
values = {'Promo2SinceWeek': 0,'Promo2SinceYear': 0, 'PromoInterval': 0}
df = df.fillna(value=values)
df
```

```python
df.isnull().sum()
```

#### Competition data

Since **CompetitionDistance** dictates whether the store is in close proximity with another competitor, **CompetitionOpenSinceMonth** and **CompetitionOpenSinceYear** should be imputed as well due to MAR case.

However, we will only use 0 as the default value for NaN values for CompetitionOpenSinceMonth and CompetitionOpenSinceYear if CompetitionDistance is NaN.

```python
cols = ['CompetitionDistance', 'CompetitionOpenSinceMonth','CompetitionOpenSinceYear']
competition = df[cols]
competition
```

```python
df.loc[df.CompetitionDistance.isna(), 'CompetitionOpenSinceMonth'] = df.CompetitionOpenSinceMonth.fillna(0)
df.loc[df.CompetitionDistance.isna(), 'CompetitionOpenSinceYear'] = df.CompetitionOpenSinceYear.fillna(0)
df['CompetitionDistance'] = df.CompetitionDistance.fillna(0)
```

```python
df.info()
```

```python
df.isnull().sum()
```

In the case of remaining missing values, we can use the earliest open dates of the **Store** column, January 2013 which coincidentally is the earliest date of all the stores

```python
df['Date'] = pd.to_datetime(df['Date'])
df['SalesMonth'] = df['Date'].dt.month
df['SalesYear'] = df['Date'].dt.year
```

```python
open_year = df[(df['Store'] == 1) & df['Open'] == 1].Date.min().year
open_month = df[(df['Store'] == 1) & df['Open'] == 1].Date.min().month
df.drop(columns=['Date'], inplace=True)

values = {'CompetitionOpenSinceMonth': open_month,'CompetitionOpenSinceYear': open_year}
df = df.fillna(value=values)
df
```

## Modelling
Based on the dataset, this problem is framed as a regression task ergo we will be using models appropriate for the task. Starting from Linear Regression models up to Support Vector Machines.

But before we dive into modelling, let us first split the dataset into the training and test set, scale the training set to allow them to work in the same range, then train the model.

### Train Test Split

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns='Sales')
y = df[['Sales']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X.columns)
```

### Training

```python
from sklearn.linear_model import LinearRegression, Ridge, SGDRegressor, HuberRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
lr_y_pred = linear_regression.predict(X_test)

forest = RandomForestRegressor(max_depth=2, random_state=0)
forest.fit(X_train, y_train)
forest_pred = forest.predict(X_test)

xgb = XGBRegressor()
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)
```

### Metrics

We shall be using R Square and RMSE.
1. R-Squared - a simple metric to determine how much the model fits the data
2. Root Mean Squared Error (RMSE) - a way to measure how well a predictive model is making predictions compared to the actual values of the target variable

References: 
- [Regression Model Metrics](https://towardsdatascience.com/what-are-the-best-metrics-to-evaluate-your-regression-model-418ca481755b)

```python
import math
from sklearn.metrics import mean_squared_error, r2_score

print('\nLinear Regression')
print('-----------------')
print(f'Mean Squared Error: {mean_squared_error(y_test, lr_y_pred)}')
print(f'Root Mean Squared Error: {math.sqrt(mean_squared_error(y_test, lr_y_pred))}')
print(f'R-Squared: {r2_score(y_test, lr_y_pred)}')

print('\nRandom Forest Regressor')
print('-----------------')
print(f'Mean Squared Error: {mean_squared_error(y_test, forest_pred)}')
print(f'Root Mean Squared Error: {math.sqrt(mean_squared_error(y_test, forest_pred))}')
print(f'R-Squared: {r2_score(y_test, forest_pred)}')

print('\nXGB Regressor')
print('-----------------')
print(f'Mean Squared Error: {mean_squared_error(y_test, xgb_pred)}')
print(f'Root Mean Squared Error: {math.sqrt(mean_squared_error(y_test, xgb_pred))}')
print(f'R-Squared: {r2_score(y_test, xgb_pred)}')
```

Based on the results, we will select XGB as our predictive model as it produces the highest R-Squared value and the lowest RMSE value among the other evaluated models

## Deploy to Model Store

In a practical setting, it is ideal to version models for the following reasons:
1. Artificat Management
2. Baseline comparisons

Let us create a makeshift Model Store to version and to store the models for deployment

```python
import glob
import shutil
import os
from joblib import dump

def get_current_model() -> str:
    path = r"../app/model_store/*.joblib"
    try:
        return glob.glob(path)[0]
    except:
        return ""

def move_current_model_to_archive():
    try:
        src_path = get_current_model()
        dst_path = r"../app/model_store/archive"
        shutil.move(src_path, dst_path)
    except:
        return
    
def get_number_of_models_in_archive() -> int:
    count = 0
    dir_path = r"../app/model_store/archive"
    for path in os.scandir(dir_path):
        if path.is_file():
            count += 1
    return count

def save_to_model_store(model):
    get_current_model()
    move_current_model_to_archive()
    model_num = get_number_of_models_in_archive() + 1
    dump(model, f"../app/model_store/sales_prediction_{model_num}.joblib")
```

```python
save_to_model_store(xgb)
```
